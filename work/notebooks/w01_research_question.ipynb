## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
print("""### Research Question

Which content pages should be prioritized for review so that content teams can spend their limited time on the pages with the greatest opportunity for improvement?

### Unit of Analysis

One row represents one content page.

### Output

A ranked list of pages with a refresh opportunity score and supporting reason codes explaining why each page was recommended.

### Decision

The decision is which pages should be reviewed first.

### Action

A content manager or SEO specialist reviews the recommended pages and decides whether to refresh, expand, improve metadata, monitor, or remove the content.

### Cost of a Wrong Recommendation

If a low-value page is recommended, valuable time is wasted reviewing content that does not need attention.

If an important declining page is missed, the organization may continue losing traffic, visibility, and potential customers.

### Why Data or Machine Learning Helps

Large websites may contain thousands of pages, making manual review difficult. Machine learning can analyze multiple signals such as impressions, clicks, content age, engagement, and traffic trends together to prioritize pages that are most likely to benefit from review. The final decision is still made by a human.""")
```
